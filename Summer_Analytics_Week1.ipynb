```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = pd.read_csv("Cars.csv")

# Step 3: Handle Missing Values
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)

# Step 4: Basic Stats
print(df.describe())

# Step 5: Pairplot
sns.set(style="ticks", color_codes=True)
sns.pairplot(df[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']])
plt.show()

# Step 6: Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Step 7: MPG vs Model Year
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()
plt.figure(figsize=(10, 5))
sns.lineplot(x=avg_mpg_by_year.index, y=avg_mpg_by_year.values, marker='o')
plt.title("Average MPG by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 8: MPG by Origin
avg_mpg_by_origin = df.groupby('origin')['mpg'].mean()
plt.figure(figsize=(6, 4))
sns.barplot(x=avg_mpg_by_origin.index, y=avg_mpg_by_origin.values, palette='viridis')
plt.title("Average MPG by Origin")
plt.ylabel("Average MPG")
plt.xlabel("Origin")
plt.tight_layout()
plt.show()

# Step 9: Car Count by Origin
plt.figure(figsize=(6, 4))
sns.countplot(x='origin', data=df, palette='Set2')
plt.title("Number of Cars by Origin")
plt.ylabel("Count")
plt.xlabel("Origin")
plt.tight_layout()
plt.show()
```
